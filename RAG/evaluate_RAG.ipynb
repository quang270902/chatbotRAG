```python
from langchain.vectorstores.chroma import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
ST_MODEL_PATH = "VoVanPhuc/sup-SimCSE-VietNamese-phobert-base"
DB_PERSIST_PATH = "./drive/MyDrive/Db/vector/"
embeddings = HuggingFaceEmbeddings(model_name=ST_MODEL_PATH, model_kwargs={"device": "cuda"})
topic_vectordb = Chroma(embedding_function=embeddings,
                  persist_directory=DB_PERSIST_PATH)
```

```python
quesion = []
context = []
answer = []
```

```python
def augment_prompt(query: str):
    #quesion.append(query)
    # get top 2 results from knowledge base
    results = topic_vectordb.similarity_search(query, k=2)
    # get the text from the results
    source_knowledge = "\n".join([x.page_content for x in results])

    context.append([source_knowledge])
    # feed into an augmented prompt
    augmented_prompt = f"""Sử dụng ngữ cảnh bên dưới, trả lời câu hỏi.

    Contexts:
    {source_knowledge}

    Query: {query}"""
    return augmented_prompt
```
