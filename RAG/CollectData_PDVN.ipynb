```python
import pandas as pd
from bs4 import BeautifulSoup
import os
```

```python
# Thư mục chứa các tệp HTML
thu_muc = 'drive/MyDrive/Db/html/demuc/'
# Danh sách để lưu trữ nội dung HTML từng tệp
html_contents_list = []
# Lặp qua tất cả các tệp trong thư mục
for ten_tep in os.listdir(thu_muc):
    duong_dan_tep = os.path.join(thu_muc, ten_tep)
    #Kiểm tra xem đối tượng có phải là tệp không
    if os.path.isfile(duong_dan_tep) and ten_tep.endswith('.html'):
        with open(duong_dan_tep, 'r', encoding='utf-8') as file_html:
            #Đọc nội dung HTML từ tệp
            html_content = file_html.read()
    #Tạo một đối tượng BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')
        # Thêm nội dung HTML vào danh sách
    html_contents_list.append(soup)
```

```python
result_strings = []
# Lặp qua tất cả các phần tử trong danh sách
for soup in html_contents_list:
    paragraphs = soup.find('div', class_='_content')
    # Lưu nội dung của các thẻ <p> vào một mảng chuỗi
    content_text = '\n'.join([p.text for p in paragraphs.find_all('p')])
    # Loại bỏ các phần tử rỗng từ mảng
    content_text = content_text.strip().split('\n3')
    result_strings.extend(content_text)
```

```python
import re
for i in range(len(result_strings)):
  if re.match(r'^Chương (?:[IVXLC]+|\d+)', result_strings[i]):
    if text != '':
      data.append(text)
      check = []
      text = ''
  elif re.match(r'^Mục \d', result_strings[i]):
    if text != '':
      data.append(text)
      check = []
      text = ''
  elif re.match(r'^Tiểu mục \d', result_strings[i]):
    if text != '':
      data.append(text)
      check = []
      text = ''
  elif re.match(r'^Điều \d+\..+', result_strings[i]):
    if text != '':
      data.append(text)
      text = ''
      check = []
        #text += result_strings[i] + '\n'
  if result_strings[i] not in check:
    check.append(result_strings[i])
    text += result_strings[i] + '\n'
data.append(text)
```

```python
df = df[df['NoiDung'] != '']
```

```python
# Lọc và lấy các dòng mà cột "Điều" bắt đầu bằng chữ "Điều"
#filtered_df = df[df['NoiDung'].str.startswith('Điều')]
# Lọc các dòng trong DataFrame mà cột 'NoiDung' bắt đầu bằng 'Điều ' và tiếp theo là một kí tự số
filtered_df = df[df['NoiDung'].str.match(r'^Điều \d')]
```

```python
print(len(filtered_df))
```

```python
filtered_df['Dieu'] = df['NoiDung'].str.split('\n').str[0]
```

```python
#Loại bỏ Điều.......
filtered_df['NoiDung'] = filtered_df['NoiDung'].str.replace(r'^.*?\n', '', regex=True)
```

```python
# Lọc các dòng có giá trị cột "NoiDung" bắt đầu bằng dấu "..."
filtered_df = filtered_df[~filtered_df['NoiDung'].str.startswith('...')]
```

```python
filtered_df.drop_duplicates(inplace=True)
filtered_df = filtered_df[filtered_df['NoiDung'].str.len() > filtered_df['NoiDung'].str.len().min()]
filtered_df.drop_duplicates(subset=['Dieu'], keep='first', inplace=True)
```

```python
# Tính độ dài chuỗi lớn nhất và nhỏ nhất trong cột noidung
max_length = filtered_df['NoiDung'].str.len().max()
min_length = filtered_df['NoiDung'].str.len().min()

print("Độ dài chuỗi lớn nhất trong cột noidung:", max_length)
print("Độ dài chuỗi nhỏ nhất trong cột noidung:", min_length)
```
